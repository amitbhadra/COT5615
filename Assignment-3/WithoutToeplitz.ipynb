```julia
function convolute(a,b)
    len_a = length(a)
    len_b = length(b)
    len_c = len_a + len_b -1
    c = Array{Float64}(undef, len_c)
    for k=1:len_c 
        sum = 0
        for j=1:len_b
            i = k+1-j
            if(i>0 && i<=len_a)
                sum+=a[i]*b[j]
            end
        end
        c[k] = sum
    end
    return c
end

u = Array{Int64}(undef, 50) 
for i=1:50 
    x = rand(Int,1) 
    if(x[1]%2==0) 
        u[i]=1 
    else 
        u[i]=-1 
    end 
end

c = [1,0.7,-0.3]
y = convolute(u, c)
using Plots
pyplot()
plot([y])

```
